```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage to crawl
url = "https://www.myhora.com/%E0%B8%AB%E0%B8%A7%E0%B8%A2/%E0%B8%AA%E0%B8%96%E0%B8%B4%E0%B8%95%E0%B8%B4%E0%B8%AB%E0%B8%A7%E0%B8%A2-%E0%B8%A2%E0%B9%89%E0%B8%AD%E0%B8%99%E0%B8%AB%E0%B8%A5%E0%B8%B1%E0%B8%87-20-%E0%B8%9B%E0%B8%B5.aspx?mode=year-range&value=20"

# Send a GET request to the webpage
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the webpage content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the table with the specified ID
    table = soup.find('table', id='dl_lottery_stats_list')
    list_data = []
    # Check if the table was found
    if table:
        # Extract table headers
        headers = [header.text.strip() for header in table.find_all('th')]
        print("Headers:", headers)
        
        # Extract table rows
        rows = table.find_all('tr')
        
        # Loop through rows and extract data
        for row in rows:
            columns = row.find_all('td')
            data = [column.text.strip() for column in columns]
            if data:  # Skip rows without data
                list_data.append(data)
                print("Row data:", data)
    else:
        print("Table with ID 'dl_lottery_stats_list' not found.")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

*** MAKE CSV

```python
import pandas as pd

# Example list_data (each item is a row of data)

# Initialize a list to hold the reformatted data
reformatted_data = []

# Loop through the list_data and reformat each row
for row in list_data:
    # Split the row data by \n and strip extra whitespace
    data = [item.strip() for item in row[0].split('\n') if item.strip()]
    
    # Reformat the data
    date = ' '.join(data[1:4])
    first_prize = data[5]
    two_digit_top = data[6]
    three_digit_top = data[7]
    two_digit_bottom = data[8]
    
    # Split three_digit numbers and format them
    three_digit_numbers = data[9].split()
    three_digit_front = ' '.join(three_digit_numbers[:2])
    three_digit_back = ' '.join(three_digit_numbers[2:])
    
    # Append the reformatted row to the data list
    reformatted_data.append([date, first_prize, two_digit_top, three_digit_top, two_digit_bottom, three_digit_front, three_digit_back])

# Create a pandas DataFrame from the reformatted data
df = pd.DataFrame(reformatted_data, columns=['วันที่', 'รางวัลที่ 1', '2 ตัวบน', '3 ตัวบน', '2 ตัวล่าง', '3 ตัวหน้า', '3 ตัวล่าง'])

# Print the DataFrame
print(df)

```
